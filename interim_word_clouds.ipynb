```python
!pip install wordcloud

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Refined Control keywords
control_keywords = """
faint muted foggy khaki beige sleepy relaxed peaceful minimal lines circles spirals squiggles
zoning out cloudy central soft light white visual field daydream gentle horizons flickering subdued
"""

# Intervention keywords
intervention_keywords = """
vivid colorful immersive kaleidoscopic spirals tunnels galaxies rainbows lava waterfall forest
rotation motion movement intense explosive fluorescent pink blue green orange purple
nature cosmic waves flow ascending universe symbolism central fixation vivid detail faces hands feathers stairs
"""

# Generate word clouds
control_wc = WordCloud(
    width=1200,
    height=600,
    background_color="white",
    collocations=False
).generate(control_keywords)

intervention_wc = WordCloud(
    width=1200,
    height=600,
    background_color="white",
    collocations=False
).generate(intervention_keywords)

# Control figure
plt.figure(figsize=(12, 6))
plt.imshow(control_wc, interpolation="bilinear")
plt.axis("off")
plt.title("Control Condition - Word Cloud", fontsize=24)
plt.tight_layout()
plt.savefig("control_wordcloud.png", dpi=300, bbox_inches="tight")
plt.show()

# Intervention figure
plt.figure(figsize=(12, 6))
plt.imshow(intervention_wc, interpolation="bilinear")
plt.axis("off")
plt.title("Intervention Condition - Word Cloud", fontsize=24)
plt.tight_layout()
plt.savefig("intervention_wordcloud.png", dpi=300, bbox_inches="tight")
plt.show()
```
